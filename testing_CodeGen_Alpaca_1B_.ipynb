````python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
````

````python
model_id = "key-life/codegen-alpaca-1b"
````

````python
# Load model & tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto")
````

````python
# Example prompt
prompt = "### Instruction:\nWrite a C++ function to check if a number is even.\n\n### Response:\n"
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

````

````python
# Generate code
outputs = model.generate(**inputs, max_new_tokens=128)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
````

````python
decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Try to extract code inside triple backticks
import re
code_block = re.findall(r"```(?:cpp|c\+\+|C\+\+)?(.*?)```", decoded, re.DOTALL)

if code_block:
    # Clean & strip extra spaces
    response = code_block[0].strip()
else:
    # Fallback: split at Response marker
    response = decoded.split("### Response:")[-1].strip()

print(response)
````
